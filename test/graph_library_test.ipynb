```python
from graph_tool import Graph, draw
from numpy.random import randint

g = Graph(directed=False)
g.add_vertex(100)

# insert some random links
for s,t in zip(randint(0, 100, 100), randint(0, 100, 100)):
    g.add_edge(g.vertex(s), g.vertex(t))

draw.graph_draw(g)
```

```python
vertex_property = g.new_vertex_property("int")
for i in range(g.num_vertices()):
    vertex_property[g.vertex(i)] = i*10

draw.graph_draw(g, vertex_text=vertex_property)
# draw.graph_draw(g, pos=..., vprops=...)  vprops = dictionary of vertex props with settings for drawing each vertex
```
